#Binary Image

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 5x5 binary image with 0s and 1s
binary_image = np.array([[0, 1, 1, 0, 0],
                         [0, 1, 0, 1, 0],
                         [1, 0, 0, 0, 1],
                         [0, 1, 0, 1, 0],
                         [0, 0, 1, 1, 0]])

# Display the image
plt.imshow(binary_image, cmap='gray')
plt.show()
```

#Grayscale Image

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 5x5 grayscale image with values from 0 to 255
grayscale_image = np.array([[0, 50, 100, 150, 200],
                            [20, 70, 120, 170, 220],
                            [40, 90, 140, 190, 240],
                            [60, 110, 160, 210, 255],
                            [80, 130, 180, 230, 50]])

# Display the grayscale image
plt.imshow(grayscale_image, cmap='gray')
plt.show()
```

#Color Images

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 5x5 color image (RGB)
color_image = np.array([[[255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0], [255, 0, 255]],
                        [[0, 255, 255], [128, 128, 128], [255, 128, 0], [0, 128, 255], [255, 255, 255]],
                        [[64, 64, 64], [192, 192, 192], [255, 64, 128], [128, 255, 64], [64, 128, 255]],
                        [[255, 128, 255], [128, 64, 255], [64, 255, 128], [255, 0, 128], [128, 255, 0]],
                        [[128, 0, 255], [255, 64, 64], [0, 128, 128], [64, 255, 64], [0, 64, 255]]])

# Display the color image
plt.imshow(color_image)
plt.show()
```

#Multispectral Images

```python
import numpy as np

# Create a 4-band multispectral image (5x5 pixels)
multispectral_image = np.random.randint(0, 256, (5, 5, 4))

# Each 'layer' corresponds to a different spectral band (here simulated with random values)
print("Band 1 (Red):\n", multispectral_image[:, :, 0])
print("Band 2 (Green):\n", multispectral_image[:, :, 1])
print("Band 3 (Blue):\n", multispectral_image[:, :, 2])
print("Band 4 (Infrared):\n", multispectral_image[:, :, 3])

plt.imshow(multispectral_image)
plt.show()
```
