```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/download.png')

# Define the top-left and bottom-right corners of the rectangle
start_point = (50, 50)  # Example starting point (x, y)

print("Image shape:", image.shape)  # Print image shape

# Define ending point based on image size
end_point = (image.shape[1]-50, image.shape[0]-50)  # Ending point (width-50, height-50)

# Define the color of the rectangle (BGR format)
color = (0, 255, 0)  # Green color

# Define the thickness of the rectangle
thickness = 2  # Thickness of the rectangle border

# Draw the rectangle on the image
cv2.rectangle(image, start_point, end_point, color, thickness)

# Convert the image from BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with the rectangle
plt.figure(figsize=(8, 6))
plt.imshow(image_rgb)
plt.title('Image with Rectangle and Grid')

# Add grid lines
plt.grid(color='white', linestyle='--', linewidth=0.5)  # Set grid color, style, and width

# Show ticks for both axes
plt.xticks(range(0, image.shape[1], 50))  # Set x ticks every 50 pixels
plt.yticks(range(0, image.shape[0], 50))  # Set y ticks every 50 pixels

plt.axis('on')  # Show axes
plt.show()
```
